```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# 1. Load the dataset
# Note: 'latin-1' encoding is standard for this specific dataset
df = pd.read_csv('spam.csv', encoding='latin-1')

# Keep only the necessary columns and rename them for clarity
df = df[['v1', 'v2']]
df.columns = ['Label', 'Text']

# Convert labels to binary numbers (Spam = 1, Ham = 0)
df['Label'] = df['Label'].map({'spam': 1, 'ham': 0})

# 2. Split data into Training and Testing sets
X = df['Text']
Y = df['Label']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 3. Vectorization (Text to Numbers)
vectorizer = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)

# 4. Save the Vectorizer
with open('vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

print("Data preprocessing complete. Vectorizer saved as vectorizer.pkl")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score

# 1. Train Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train_features, Y_train)
lr_pred = lr_model.predict(X_test_features)

print("--- Logistic Regression ---")
print("Accuracy:", accuracy_score(Y_test, lr_pred))
print("Precision:", precision_score(Y_test, lr_pred))

# 2. Train Naive Bayes (Usually better for text)
nb_model = MultinomialNB()
nb_model.fit(X_train_features, Y_train)
nb_pred = nb_model.predict(X_test_features)

print("\n--- Naive Bayes ---")
print("Accuracy:", accuracy_score(Y_test, nb_pred))
print("Precision:", precision_score(Y_test, nb_pred))

# 3. Save the Naive Bayes model (as it typically has higher precision for spam)
with open('model.pkl', 'wb') as f:
    pickle.dump(nb_model, f)

print("\nModel training complete. Best model saved as model.pkl")
```
